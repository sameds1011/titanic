```python
##eksik verileri tanımlayaıcı fonksyion hazırlayalaım

def missing_value(df,na_name=False):
  ##İçerisinde boş veri olan kolon adlarını topladım
  na_columns=[col for col in df.columns if df[col].isnull().sum()>0]
    ##Kaç tane boş veri var buldum sıralaım
  n_miss=df[na_columns].isnull().sum().sort_values(ascending=False)
    ##Kaç tane boş veri oransal buldum sıralaım
  ratio=(df[na_columns].isnull().sum()/len(df)*100).sort_values(ascending=False)
    ##oransal ve sayısal birleştirdim
  missing_df=pd.concat([n_miss,ratio],axis=1,keys=["sayısal","oransal"])
  return missing_df
```

```python
##Numerik veriler : yaş, km bilgisi gibi
## Katgeorik veriler: Kurtuldu mu,liman bilgisi,cinsiyet
##Kardinal veri : kategori sayısı çok yüksek
def grab_col_names(df,cat_th=10,car_th=20):
#dtypes="o" Objectliği sorgular
#Object olan veriler otamatik olarak kategoriktir
  cat_cols=[col for col in df.columns if df[col].dtypes=="O"]
##Numerik gözüken ama kategorik olan verileri topladım
  num_but_cat=[col for col in df.columns if df[col].nunique()<cat_th and df[col].dtypes!="O"]
 ##Kategorik gibi gözüküp kardinal olan verileri topladım
  cat_but_car=[ col for col in df.columns if df[col].nunique()>car_th and df[col].dtypes=="O"]
##Kategorik verileri toparladım
  cat_cols=cat_cols+num_but_cat
  ##Kardinal verileri listeden çıkar
  cat_cols=[col for col in cat_cols if col not in cat_but_car]
  ##Numerik veri listesini oluşturucam
  num_cols= [col for col in df.columns if  df[col].dtypes!="O"]
  ###Numerik olarak gözüküp kategorik olanlar var onları numerik listesinden çıkaralım
  num_cols=[col for col in num_cols if col not in num_but_cat]

  print(f"Kaç Satır var: {df.shape[0]}")
  print(f"Kaç değişken(Sutün) var: {df.shape[1]}")
  print(f"Kategorik Veriler : {len(cat_cols)}")
  print(f"Numerik Veriler : {len(num_cols)}")
  print(f"Kardinal Veriler : {len(cat_but_car)}")
  print(f"Numerik Gözüken Kategorik Veriler : {len(num_but_cat)}")
  return cat_cols,num_cols,cat_but_car
```

```python
kategorikler,numerikler,kardinaller=grab_col_names(df)
```

```python
print("Kategorikler :  " ,kategorikler)
print("Numerikler :  " ,numerikler)
print("Kardinaler :  " ,kardinaller)
```

```python
def missing_vs_target(df,target,na_columns):
  temp_df=df.copy()
  for col in na_columns:
    temp_df[col +"_NA_FLAG"]=np.where(temp_df[col].isnull(),1,0)
  na_flags=temp_df.loc[:,temp_df.columns.str.contains("_NA_")].columns
  for col in na_flags:
    print(pd.DataFrame({"Ortalama": temp_df.groupby(col)[target].mean(),
                        "Saydır":temp_df.groupby(col)[target].count()}))
```

```python
missing_vs_target(df,"survived", ["age","embarked"])
```
